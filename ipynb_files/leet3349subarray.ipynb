```python
class Solution(object):

    def runSolution(self, nums, k):
            lis = set()
            subarr = []
            for i in range(len(nums)-2*k):
                a = i 
                b = i + k -1
                print(nums[a:a+k], nums[b:b+k])
                for d in range(len(nums[a:a+k])):
                    if nums[d+1] > nums[d]:
                        lis.add(nums[d])
                        lis.add(nums[d+1])
                    if len(lis) == k:
                        subarr.append(list(lis))
                        lis = set()
                        break
                    
                for d in range(len(nums[b:b+k])):
                    if nums[d+1] > nums[d]:
                        lis.add(nums[d])
                        lis.add(nums[d+1])
                    if len(lis) == k:
                        print(lis)
                        subarr.append(list(lis))
                        lis = set()
                        break
                    
            if len(subarr) >= 2:
                print(lis)
                print(subarr)
                return True
            else:
                print(lis)
                print(subarr)
                return False

    def hasIncreasingSubarrays(self, nums, k):

        if len(nums) - k >= k:
            if k == 1 : return True           
            return self.runSolution(nums, k)
        else:
            return False

if __name__ == "__main__":
    
    array = [5,8,-2,-1]
    k = 2
    r = Solution()
    
    val = r.hasIncreasingSubarrays(array, k)
    print(val)  
    
```
